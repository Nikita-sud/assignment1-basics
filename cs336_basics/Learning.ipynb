```python
import regex as re

text = "I love AI"
pattern = r"love"
match = re.search(pattern, text)

print(match.group()) # Выведет: "love"
print(match.start()) # Выведет: 2 (потому что 'I' - 0, ' ' - 1, 'l' - 2)
print(match.end())   # Выведет: 6
print(match.span())  # Выведет: (2, 6)
print(match)
byte_tokens = [bytes([b]) for b in text.encode("utf-8")]
print(byte_tokens)
```
